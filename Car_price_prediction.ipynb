```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python

def plotdata(data, col_name, col_type):
    if col_type == 'object':
        plt.figure(figsize=(15,3))
        sns.countplot(x=col_name, data=data, palette='YlGnBu')
        plt.show()
```

```python
for i in car_data.columns:
    print(i)
    plotdata(car_data, i, car_data[i].dtype)
```

```python
car_data.Car_Name.unique()
```

```python
brand = lambda x: x.split(' ')[0]
```

```python
car_data['Car_Name'] = car_data['Car_Name'].apply(brand)
```

```python
for i in car_data.columns:
    print(i)
    plotdata(car_data, i, car_data[i].dtype)
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()

def labelTrandformer(col_name, data, col_type):
    if col_type == 'object':
        data[col_name+"_encoded"]  = label_encoder.fit_transform(data[col_name])
```

```python
for i in car_data.columns:
    labelTrandformer(i, car_data,car_data[i].dtype)
car_data.head()
```

```python
X = car_data[['Present_Price', 'Kms_Driven', 'Owner', 'Year', 'Car_Name_encoded', 'Fuel_Type_encoded', 'Seller_Type_encoded', 'Transmission_encoded']]
Y = car_data[['Selling_Price']]
```

```python
X.head()
```

```python
Y.head()
```

```python
X.shape, Y.shape
```

```python
sns.pairplot(pd.concat([X,Y], axis=1))
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=1)

print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.metrics import r2_score
```

```python
linear_regression = LinearRegression()
linear_regression.fit(x_train, y_train)
linear_regression_predictions = linear_regression.predict(x_test)
linear_regression_r2_score = r2_score(y_test, linear_regression_predictions)
```

```python
linear_regression_r2_score
```

```python
list(zip(y_test.values, linear_regression_predictions))
```

```python
Lasso_regression = Lasso()
Lasso_regression.fit(x_train, y_train)
Lasso_regression_predictions = Lasso_regression.predict(x_test)
Lasso_regression_r2_score = r2_score(y_test, Lasso_regression_predictions)
```

```python
Lasso_regression_r2_score
```

```python
list(zip(y_test.values, Lasso_regression_predictions))
```

```python
Ridge_regression = Ridge()
Ridge_regression.fit(x_train, y_train)
Ridge_regression_predictions = Ridge_regression.predict(x_test)
Ridge_regression_r2_score = r2_score(y_test, Ridge_regression_predictions)
```

```python
Ridge_regression_r2_score
```

```python
list(zip(y_test.values, Ridge_regression_predictions))
```
